# DATA CLEANING

```python
df.head()
```

```python
# clean the dataset of unnecessary columns like instant as its just an index and also the dteday as the day, month and year columns are already present along with casual and registered as it is validated in count

df.drop(['instant','dteday','holiday','casual','registered'], axis=1 , inplace=True)
df.head()
```

```python
#Renaming the columns
df.rename(columns = {'yr':'year','mnth':'month','hum':'humidity','cnt':'count'}, inplace = True) 
df.head()
```

```python
# referring the data dictionary we can replace the numbers with the actual meaning for season,month etc..

df['season'].replace({1:'spring', 2:'summer', 3:'fall', 4:'winter'}, inplace=True)
df.head()
```

```python
df['weathersit'].replace({1:"Clear/Few clouds",2:"Misty clouds",3:"Light snow/rain ",4:'Heavy Rain/thunderstorm/snow'},inplace=True)
df.head()
```

```python
df['weekday'].replace({0:"Mon",1:"Tue",2:"Wed",3:"Thu",4:"Fri",5:"Sat",6:"Sun"},inplace=True)
df.head()
```

```python
df['month'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'},inplace=True)
df.head()
```

# Dummy Variables

```python
#create dummy varibales for categorical columns

season_dum=pd.get_dummies(df['season'],drop_first=True)
month_dum=pd.get_dummies(df['month'],drop_first=True)
weekday_dum=pd.get_dummies(df['weekday'],drop_first=True)
weathersit_dum=pd.get_dummies(df['weathersit'],drop_first=True)
```

```python
df.head()
```

```python
# add these dummies to the data set and delete the parent columns

df=pd.concat([df,season_dum],axis=1)
df=pd.concat([df,month_dum],axis=1)
df=pd.concat([df,weekday_dum],axis=1)
df=pd.concat([df,weathersit_dum],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
# now we can delete the columns season,month,weekday,weathersit

df.drop(['season'],axis=1 , inplace=True)
df.drop(['month'],axis=1 , inplace=True)
df.drop(['weekday'],axis=1 , inplace=True)
df.drop(['weathersit'],axis=1 , inplace=True)
```

```python
df.head()
```

```python
# check correlation between the columns and refer 

plt.figure(figsize=(25, 20))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

**Seen that year and temp have a high correlation with count and negatively with spring**

### Splitting the data into training and test data

```python
# let's take 70% data in our training set and the rest for test

from sklearn.model_selection import train_test_split

df_train , df_test = train_test_split(df, train_size=0.7 , random_state=100)
```

```python
df_train.head()
```

### Rescaling

```python
# Scaling using minmaxscaler

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
df.columns
```

```python
# fit-transform on numerical variables 

nums=['temp','humidity','windspeed','count']
df_train[nums]=scaler.fit_transform(df_train[nums])
df_train[nums].head()
```

## Creating variables X and Y

```python
# store only count in y_train as it is the independent variable here

y_train=df_train.pop('count')
x_train=df_train
```

```python
x_train.head()
```

## RFE

```python
#import RFE and Linear regression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# starting the model with 20 vars

lm=LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,20)
rfe=rfe.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

**create a dataset with supported columns as given by rfe**

```python
# storing rfe supported vars 

supp=x_train.columns[rfe.support_].sort_values()
supp
```

```python
#listing the rfe support with false value

x_train.columns[~rfe.support_]
```

```python
#creating a dataset now to work our models upon

x_train_rfe = x_train[supp]
x_train_rfe.head()
```

## Build model using statsmodels

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# add constant

x_train_rfe=sm.add_constant(x_train_rfe)

# to check for VIF values

vif = pd.DataFrame()
X = x_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
